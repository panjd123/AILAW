```python
import pandas as pd
from tqdm import tqdm
import jieba
import jieba.analyse

df = pd.read_csv('res.csv')
```

```python
df = df.sample(1000)
idx = df['罪名']=='受贿'
x_train = df['事实认定'].values
x_test = df[idx]['事实认定'].values

def tf(text):
    words = jieba.cut(text)
    d = {}
    for word in words:
        if word in d:
            d[word] += 1
        else:
            d[word] = 1
    return d

def get_idf(docs):
    d = {}
    for doc in tqdm(docs):
        words = set(jieba.cut(doc))
        for word in words:
            if word in d:
                d[word] += 1
            else:
                d[word] = 1
                
    for word in d:
        d[word] = len(docs) / d[word]
    return d

def tfidf(text, idf):
    
    if isinstance(text,str):
        d = tf(text)
    else:
        d={}
        for t in tqdm(text):
            td = tf(t)
            for word in td:
                if word in d:
                    d[word]+=td[word]
                else:
                    d[word]=td[word]
    for word in d:
        d[word] *= idf[word]
    return sorted(d.items(), key=lambda x: x[1], reverse=True)

# idf = get_idf(x_train)
# tfidf_test = tfidf(x_test, idf)
# tfidf_test
```
