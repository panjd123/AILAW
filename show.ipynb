```python
import pandas as pd
import numpy as np
import re
df = pd.read_json('./sample_p4.txt',encoding='gb18030',lines=True)
df.dropna(axis=0, subset=['content'], inplace=True)

fact =[]
def get_money(t:str):
    ret = []
    pattern = [["追缴赃款"], ["违法所得"], ["所得赃款"]]
    for patn in pattern:
        m = re.findall(f"(?<={patn}).*元",t,re.M)
        ret.extend(m)
    
    m = re.findall(f"(?<=收受的).*赃款依法予以没收",t,re.M)
    ret.extend(m)
    
    return ret
    
for idx,row in df.iterrows():
    text = row['content']
    bg = text.find("经审理查明")
    ed = text.find("本院认为")
    text = text[bg+6:ed]
    fact.append(text)
    
df.insert(2,'fact',fact)

money = []
cnt = 1

for idx,row in df.iterrows():
    if(row['content'].find("我院决定将该案指定")!=-1):
        continue
    
    m = get_money(row['content'])
    if(len(m)==0):
        cnt+=1
    money.append(m)
    
print(cnt)
```
